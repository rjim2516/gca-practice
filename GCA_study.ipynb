```python
class Solution:
    def coinChange(self, coins: List[int], amount: int) -> int:
        # Initialize dp array with amount+1 (as a stand-in for infinity)
        dp = [amount + 1] * (amount + 1)
        dp[0] = 0  # Base case: 0 coins needed to make amount 0

        # Build up the dp array
        for i in range(1, amount + 1):
            for coin in coins:
                if i - coin >= 0:
                    dp[i] = min(dp[i], dp[i - coin] + 1)

        # If we never updated dp[amount], it's impossible to make that amount
        return dp[amount] if dp[amount] != amount + 1 else -1
```

```python
def coinChange(coins, amount):
    # Step 1: Initialize the DP array
    # Set each value to infinity as a placeholder for "unreachable"
    dp = [float('inf')] * (amount + 1)

    # Base case: 0 coins are needed to make amount 0
    dp[0] = 0

    # Step 2: Build up the solution for every amount from 1 to amount
    for a in range(1, amount + 1):
        for coin in coins:
            if a - coin >= 0:
                dp[a] = min(dp[a], 1 + dp[a - coin])

    # Step 3: Return the result
    # If we never updated dp[amount], it means it's unreachable
    return dp[amount] if dp[amount] != float('inf') else -1


coins = [1, 2, 5]
amount = 11
print(coinChange(coins, amount))  # Output: 3
```

```python
def twoSum(nums, target):
  for i in range(len(nums)):
    for j in range(i + 1, len(nums)):
      if nums[i] + nums[j] == target:
        return [i,j]

print(twoSum([2, 7, 11, 15], 9))


```

```python
#Count Positive, Negative, and Zero

def count_numbers(nums):
  positive = 0
  negative = 0
  zero = 0
  for num in nums:
    if num > 0:
      positive += 1
    elif num < 0:
      negative += 1
    else:
      zero += 1
  return positive, negative, zero

print(count_numbers([-3, 0, 1, 2, -1, 0]))
```

```python
#Define first duplicate

def first_duplicate(nums):
  for i in range(len(nums)):
    for j in range(i + 1, len(nums)):
      if nums[i] == nums[j]:
        return nums[i]
  return -1

print(first_duplicate([2, 1, 3, 7, 8]))


```

```python
#Define a reversed string

def reverse_string(s):
    reversed_string = ""
    for char in s:
        reversed_string = char + reversed_string
    return reversed_string

print(reverse_string("hello"))  # → "olleh"
print(reverse_string(""))       # → ""
print(reverse_string("a"))      # → "a"
```

```python
def max_num(a_list):
    if not a_list:
        return None  # or raise an error if you prefer

    max_val = a_list[0]
    for num in a_list:
        if num > max_val:
            max_val = num
    return max_val

print(max_num([4, 9, 1, 22, 5]))  # → 22
print(max_num([-3, -7, -1]))      # → -1
print(max_num([]))               # → None

```

```python
#Check for Palindrome (simple palindrome)
def is_palindrome(s):
  return s == s[::-1]

print(is_palindrome("madam"))       # → True
print(is_palindrome("racecar"))     # → True
print(is_palindrome("hello"))       # → False
```

```python
#Check for Palindrome, ignore upper and lower case & remove spaces
def is_palindrome(s):
    cleaned = s.replace(" ", "").lower()
    return cleaned == cleaned[::-1]

print(is_palindrome("madam"))                        # True
print(is_palindrome("A man a plan a canal Panama"))  # True
print(is_palindrome("hello"))                        # False
```

```python
#sum of all positive numbers

def sum_of_positives(nums):      # Use 'nums' as parameter name (plural for a list)
  total = 0                      # Start your total at 0
  for num in nums:              # Loop through each number
    if num > 0:
      total += num              # Add the positive number itself
  return total                  # Return the final total


print(sum_of_positives([1, -3, 5, 0, -2, 6]))  # Should print 12 (1 + 5 + 6)



```
